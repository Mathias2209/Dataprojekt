# Project description

### 1. What is the overall purpose of the project?

A Danish commercial workwear company operates laundry facilities that process thousands of textile items daily. Each piece of workwear contains an RFID chip that tracks its journey through the system. Recording the product type, how many days it has been in circulation, how many wash cycles it has undergone, and eventually, why it was discarded. 

The company wants to understand the lifetime of their textiles. They would like to compare how long clothing lasts across different laundry facilities and between different suppliers. For example: Does a t-shirt from Supplier A last longer than one from Supplier B?

The main purpose of our project is to:
- **Assess data quality:** What patterns exist in the data? Are there anomalies, missing values, or
inconsistencies that suggest problems with the collection process?
- **Explore whatâ€™s possible:** What questions can this data reliably answer? What questions
remain out of reach given current data limitations?
- **Analyze the lifetime:** By identifying patterns in wear and tear and discards we can understand why and when textiles gets discarded, we can help optimize ressources and promote a more sustainable textile industry through extended product life.
- **Visualize and communicate:** Create clear visualizations that reveal patterns, both expected
and unexpected.
- **Recommend improvements:** Based on our findings, how could the company improve their
data collection to enable more robust analyses in the future?


### 2. What is the academic problem or research question?

How can we precisely model the real lifespan for industrial textiles, when the data can contain, wrong administrative registrations and some of the RFID chips is reused?


What influence does administrative reasons for discard (e.g. "discontinued model") against physical reasons (e.g. wear and tear) have on the statistical lifespan model?

Is there any significant difference in lifetime between textiletype / product type and number of washes?

### 5. Are there any specific challenges?

A typical facility at the company discards between 10,000 and 30,000 items per month, generating a substantial dataset. However, despite years of data collection, no systematic analysis has ever been performed.

Some of our challenges includes:
- **x:** Some of the datapoints have 0 days of circulation or number of washes
- **Quality of the data:** Differences in naming of the data files
- **Identity conflicts:** Some of the Unique codes (ui) is not unique. This occurs when a piece of clothing is discarded with one customer, and then taken back into use with another
- **Bulk-data:** Some of our observations have "Pieces of clothing per discard date" greater than 1
