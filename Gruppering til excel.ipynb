```python
# Find antallet af skjorter for hvert unikt produktnavn og gennemsnits livstid for hver unikke produktnavn
skjorte_produkt_counts = skjorte_data['Produkt - Produkt'].value_counts()

# Beregn gennemsnitslivstid for hvert unikt produktnavn
skjorte_produkt_gennemsnit = skjorte_data.groupby('Produkt - Produkt')['Dage i cirkulation'].mean()
skjorte_produkt_gennemsnit.to_excel('skjorte_produkt_gennemsnit.xlsx')

# Beregn gns antal vaske for hver unikt produktnavn
skjorte_produkt_gns_vaske = skjorte_data.groupby('Produkt - Produkt')['Total antal vask'].mean()
skjorte_produkt_gns_vaske.to_excel('skjorte_produkt_gns_vaske.xlsx')

# beregn gennemsnitslivstid og gns antal vaske samt antal skjorter for hver unikt produktnavn kassationsårsagen "alm.slid uden restværdi" og skriv det til excel
skjorte_alm_slid = skjorte_data[skjorte_data['Kassationsårsag (ui)'] == 'Alm.slid uden restværdi']
skjorte_alm_slid_gennemsnit = skjorte_alm_slid.groupby('Produkt - Produkt').agg({'Dage i cirkulation': 'mean', 'Total antal vask': 'mean', 'Produkt - Produkt': 'count'})
skjorte_alm_slid_gennemsnit.to_excel('skjorte_alm_slid_gennemsnit.xlsx')


```
