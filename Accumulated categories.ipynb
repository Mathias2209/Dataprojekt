```python
import matplotlib.pyplot as plt

from dataloader import samlet_df
from dataloader import(skjorte_data, shorts_data, bukse_data, tshirt_data, langærmet_data,
jakke_data, fleece_data, overall_data, forklæde_data, kittel_data, busseron_data,
kokkejakke_data)
df = samlet_df
```

```python
total = len(df)
Accumulated = 0

# Bukser
Accumulated += len(bukse_data)
print(f"Bukser udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# T-shirts
Accumulated += len(tshirt_data)
print(f"Bukser og T-shirts udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Jakker
Accumulated += len(jakke_data)
print(f"Ovenstående + Jakker udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Skjorter
Accumulated += len(skjorte_data)
print(f"Ovenstående + Skjorter udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Overalls
Accumulated += len(overall_data)
print(f"Ovenstående + Overalls udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Langærmet
Accumulated += len(langærmet_data)
print(f"Ovenstående + Langærmet udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Busseron
Accumulated += len(busseron_data)
print(f"Ovenstående + Busseron udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Kokkejakker
Accumulated += len(kokkejakke_data)
print(f"Ovenstående + Kokkejakker udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Fleece
Accumulated += len(fleece_data)
print(f"Ovenstående + Fleece udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")

# Forklæde
Accumulated += len(forklæde_data)
print(f"Ovenstående + Forklæde udgør: {Accumulated} svarende til {Accumulated/total:.2%} af det samlede dataset.")
```

```python
Bukser = len(bukse_data) 
Tshirts = Bukser + len(tshirt_data)
Jakker = Tshirts + len(jakke_data)
Skjorter = Jakker + len(skjorte_data)
Overalls = Skjorter + len(overall_data)
Langærmet = Overalls + len(langærmet_data)
Busseron = Langærmet + len(busseron_data)
Kokkejakker = Busseron + len(kokkejakke_data)
Fleece = Kokkejakker + len(fleece_data)
Forklæde = Fleece + len(forklæde_data)

procent = [Bukser/total*100, Tshirts/total*100, Jakker/total*100, Skjorter/total*100, Overalls/total*100, 
           Langærmet/total*100, Busseron/total*100, Kokkejakker/total*100, Fleece/total*100, Forklæde/total*100]
kategorier = ['Bukser', '+T-shirts', '+Jakker', '+Skjorter', '+Overalls', '+Langærmet', '+Busseron', '+Kokkejakker', '+Fleece', '+Forklæde']

# Plot
plt.figure(figsize=(15, 6))
plt.bar(kategorier, procent)
plt.xlabel('Kategorier')

# %
plt.yticks(range(0, 101, 10), [f"{i}%" for i in range(0, 101, 10)])
for idx, val in enumerate(procent):
    plt.text(idx, val + 1, f"{val:.1f}%", ha='center')

plt.title('Akkumulerede kategorier')
plt.show()
```
