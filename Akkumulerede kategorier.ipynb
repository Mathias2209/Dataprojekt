# Akkumulerede kategorier

Vi ønsker at undersøge fordeling af de forskellige tøj-kategorier vi har opdelt vores datasæt i. Vi beregner derfor den akkumulerede andel af kategorierne.

```python
# Beregning af akkumulerede kategorier
total = len(df)
Accumulated = 0

# Kategorier
## Bukser
Accumulated += len(bukse_data)
print(f"Bukser udgør:                       {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## T-shirts
Accumulated += len(tshirt_data)
print(f"Bukser og T-shirts udgør:           {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Jakker
Accumulated += len(jakke_data)
print(f"Ovenstående + Jakker udgør:         {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Skjorter
Accumulated += len(skjorte_data)
print(f"Ovenstående + Skjorter udgør:       {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Overalls
Accumulated += len(overall_data)
print(f"Ovenstående + Overalls udgør:       {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Langærmet
Accumulated += len(langærmet_data)
print(f"Ovenstående + Langærmet udgør:      {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Busseron
Accumulated += len(busseron_data)
print(f"Ovenstående + Busseron udgør:       {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Kokkejakker
Accumulated += len(kokkejakke_data)
print(f"Ovenstående + Kokkejakker udgør:    {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Fleece
Accumulated += len(fleece_data)
print(f"Ovenstående + Fleece udgør:         {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Forklæde
Accumulated += len(forklæde_data)
print(f"Ovenstående + Forklæde udgør:       {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Kittel
Accumulated += len(kittel_data)
print(f"Ovenstående + Kittel udgør:         {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")

## Shorts
Accumulated += len(shorts_data)
print(f"Ovenstående + Shorts udgør:         {Accumulated:,} svarende til {Accumulated/total:.2%} af det samlede datasæt.")
```

Det ses, at den akkumulerede andel stiger kraftigt i starten, og derefter flader en del ud.

For bedre at visualisere dette, plottes andelene.

## Plot

```python
# Akkumulerede kategorier (X-akse)
Bukser = len(bukse_data) 
Tshirts = Bukser + len(tshirt_data)
Jakker = Tshirts + len(jakke_data)
Skjorter = Jakker + len(skjorte_data)
Overalls = Skjorter + len(overall_data)
Langærmet = Overalls + len(langærmet_data)
Busseron = Langærmet + len(busseron_data)
Kokkejakker = Busseron + len(kokkejakke_data)
Fleece = Kokkejakker + len(fleece_data)
Forklæde = Fleece + len(forklæde_data)
Kittel = Forklæde + len(kittel_data)
Shorts = Kittel + len(shorts_data)

# Akse labels
kategorier = ['Bukser', '+T-shirts', '+Jakker', '+Skjorter', '+Overalls', '+Langærmet',
              '+Busseron', '+Kokkejakker', '+Fleece', '+Forklæde', '+Kittel', '+Shorts', '+Andet']

# Procentvis fordeling (Y-akse)
procent = [Bukser/total*100, Tshirts/total*100, Jakker/total*100, Skjorter/total*100,
        Overalls/total*100, Langærmet/total*100, Busseron/total*100, Kokkejakker/total*100, 
        Fleece/total*100, Forklæde/total*100, Kittel/total*100, Shorts/total*100, 100]


# Plot
plt.figure(figsize=(15, 6))
plt.bar(kategorier, procent)
plt.xlabel('Kategorier')

# %
plt.yticks(range(0, 101, 10), [f"{i}%" for i in range(0, 101, 10)])
for idx, val in enumerate(procent):
    plt.text(idx, val + 1, f"{val:.1f}%", ha='center')

# 70% og 90% linje
plt.axhline(y=70, color='red', linestyle='--', linewidth=1.5, alpha=0.4)
plt.axhline(y=90, color='red', linestyle='--', linewidth=1.5, alpha=0.4)

plt.title('Akkumulerede kategorier')
plt.show()
```

Vi ser at $70%$ af vores data ligger under kun $3$ kategorier: $\\$
Bukser, T-shirts & Jakker

Og at $90%$ af tøjet er fordelt på de $7$ største kategorier: $\\$
Bukser, T-shirts, Jakker, Skjorter, Overalls, Langærmet & Busseron

De resterende kategorier udgør altså en meget lille del af vores datasæt.
